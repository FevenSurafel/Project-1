```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File to Load
nasdaq_data = "Resources/Main.csv"
```

```python
# Read Data File and store into Pandas DataFrames
nasdaq_data_df = pd.read_csv(nasdaq_data)
nasdaq_data_df
```

```python
stock_data = nasdaq_data_df['Company'].unique()
stock_data
```

```python
#Calcuate the Dividend Yield then create column
nasdaq_data_df["Dividend Yield"] = (nasdaq_data_df["Ex-Dividend"] / nasdaq_data_df['Close'])

nasdaq_data_df[nasdaq_data_df["Dividend Yield"] > 0]
```

```python
grouped_stocks_df = nasdaq_data_df.groupby(["Company"])
```

```python
div_yield = grouped_stocks_df["Dividend Yield"].mean()
div_yield
```

```python
# Configure plot, figsize, title, and axis labels
figure1 = div_yield.plot(kind="bar", facecolor="purple", figsize=(8,6),
                                title="Dividend Yield for Different Stocks",
                                xlabel="Company",
                                ylabel="Dividend Yield (Mean)")

# Configure x-tick rotation
xticklabels = div_yield.index
figure1.set_xticklabels(xticklabels, rotation=45, rotation_mode="anchor", ha="right", wrap=True)

# Show plot
plt.show()

# Resize plot to display labels
plt.tight_layout()
```
