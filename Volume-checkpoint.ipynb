```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File to Load
nasdaq_data = "Resources/Main.csv"
```

```python
# Read Data File and store into Pandas DataFrames
nasdaq_data_df = pd.read_csv(nasdaq_data)
nasdaq_data_df
```

```python
# Calculate the average trading volume
average_volume = nasdaq_data_df['Volume'].mean()

# Set a threshold for unusually high activity (10x the average)
threshold = 10 * average_volume

# Filter days with trading volume above the threshold
unusually_high_volume_days = nasdaq_data_df[nasdaq_data_df['Volume'] > threshold]
unusually_high_volume_days
```

```python
stock_data = nasdaq_data_df["Company"].unique()
stock_data
```

```python
unusually_high_volume_days

nasdaq_data_df = pd.DataFrame(unusually_high_volume_days)
nasdaq_data_df['Date'] = pd.to_datetime(nasdaq_data_df['Date'])

# Configure plot, figsize, title, and axis labels
plt.figure(figsize=(10, 5))
plt.plot(nasdaq_data_df['Date'], nasdaq_data_df['Volume'], marker='o', linestyle='-')
plt.title('Unusually High Volume Days')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)

# Show plot
plt.show()
```
